# Array module

```python
import array
```

```python
# Creation
arr1 = array.array("i", [1, 2, 3, 4, 5])
arr1
```

```python
# Accessing elements
arr1[2]     # Third element
```

```python
# Traverse the array
for element in arr1:
    print(element, end=" ")
```

```python
arr1.insert(5, 20)
arr1
```

```python
# Deletion
arr1.remove(20)     # Removes 20 (only removes the first occurrence)
```

# Numpy

```python
import numpy as np
```

```python
# Creation

np1 = np.array([10, 20, 30, 40, 50])
np2 = np.array([50, 60, 70])

print(np1)
print(np2)
```

```python
# Access
np1[1]
```

```python
# Traversal
for element in np1:
    print(element)
```

```python
# Insertion
np.insert(np1, 2, 100)
```

```python
# Deletion
np.delete(np1, 4)
```

```python
# Updation
np1[0] = 250
np1
```

```python
# Sorting
np.sort(np1)
```

```python
# Merging
print(np1)
print(np2)
```

```python
combined_array = np.concatenate((np1, np2))
combined_array
```

```python
# Splitting
split1, split2 = np.array_split(combined_array, 2)
print(split1)
print(split2)
```
