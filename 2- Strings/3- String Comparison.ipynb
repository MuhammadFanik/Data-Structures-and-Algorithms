# Case sensitivity

```python
str1 = 'zoo'
str2 ='zoO'
```

```python
str1 == str2
```

```python
for ch1, ch2 in zip(str1, str2):
    print(ch1, ord(ch1), '-', ch2, ord(ch2))
```
