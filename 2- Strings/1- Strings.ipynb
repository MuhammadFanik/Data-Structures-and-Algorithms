# 1 - Strings in Python

## Ways of creating a String

```python
str1 = "This is a string in Python"
print(str1)
print(type(str1))
```

```python
str2 = "This is also a string in Python"
print(str2)
print(type(str2))
```

```python
str3 = '''This is a string within triple quotes'''
print(str3)
print(type(str3))
```

```python
str4 = """This is also a string within triple quotes"""
print(str4)
print(type(str4))
```

```python
print('Hi "John"')
```

```python
print('''This is a
multi-line string that spans
multiple lines''')
```

```python
print('This is a \
multi-line string that spans \
multiple lines')
```

## Properties of Strings

```python
str1 = "Hello"
print(id(str1))

str1 = str1 + " World"
print(str1)
print(id(str1))
```

```python
# Iterable
for ch in str1:
    print(ch, end="")
```

```python
# Slicing
str1[2:5]
```

```python
# Unicode
print(ord("a"), ord("z"))
```

```python
# Concatenation
s1 = "Data"
s2 = "Science"
print(s1 + " " + s2)
```

```python
# Repetition
s1 * 2
```

## Common String Functions

```python
str1
```

```python
print(id(str1))
```

```python
# lower
lower_str1 = str1.lower()
print(lower_str1)
print(id(lower_str1))
```

```python
# upper
str1.upper()
```

```python
# Strip
str3 = "     data    science    "
str3.strip()
```

```python
str4 = '$$$$###data science#############@@@@@'
str4.strip("$").strip("@").strip("#")
```

```python
# replace
str1 = "Data Science"
str1.replace("Science", "Analytics")
```

```python
# Split
str2 = 'Data Science and Data Engineering'
(str2.split())
```

```python
print(type(str2), type(str2.split()))
```

```python
str3 = "Data-Science-and-Data-Engineering"
str3.split("-")
```

```python
# Join
list1 = ["My", "name", "is", "Jack"]
list1
```

```python
" ".join(list1)
```
